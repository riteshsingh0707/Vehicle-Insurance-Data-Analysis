```python
categorical_cols = df.select_dtypes(include=['object']).columns
print("\n--- Unique Values in Categorical Columns ---")
for col in categorical_cols:
    print(f"{col}: {df[col].unique()}")
```

```python
# . Check for duplicates
duplicates = df.duplicated().sum()
print("Number of duplicate rows:", duplicates)
if duplicates > 0:
    df = df.drop_duplicates()
```

```python
# Select numeric columns
num_cols = df.select_dtypes(include=['int64','float64']).columns

# IQR-based outlier detection for all numeric columns
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = ((df[col] < lower) | (df[col] > upper)).sum()
    print(f"{col}: Outliers = {outliers}, Range = ({df[col].min()}, {df[col].max()})")
```

```python
#Convert float-coded categorical columns to integer
df['Region_Code'] = df['Region_Code'].astype(int)
df['Policy_Sales_Channel'] = df['Policy_Sales_Channel'].astype(int)
```

```python
#Map Vehicle_Age (ordinal category â†’ numeric code)
age_map = {'< 1 Year': 0, '1-2 Year': 1, '> 2 Years': 2}
df['Vehicle_Age_Code'] = df['Vehicle_Age'].map(age_map)
```

```python
# Age vs Response
plt.figure(figsize=(6,4))
sns.kdeplot(df[df['Response']==0]['Age'], label='No Response', shade=True)
sns.kdeplot(df[df['Response']==1]['Age'], label='Response', shade=True)
plt.title("Age vs Response")
plt.xlabel("Age")
plt.legend()
plt.show()
```

```python
# Premium vs Response
plt.figure(figsize=(6,4))
sns.boxplot(x='Response', y='Annual_Premium', data=df)
plt.title("Annual Premium vs Response")
plt.show()
```

```python
#Claim Frequencies (Factors)
# Response frequency by key features
print("Response Rate by Vehicle Damage:\n", df.groupby('Vehicle_Damage')['Response'].mean()*100)
print("\nResponse Rate by Vehicle Age:\n", df.groupby('Vehicle_Age')['Response'].mean()*100)
print("\nResponse Rate by Previously Insured:\n", df.groupby('Previously_Insured')['Response'].mean()*100)
```

```python
#Gender Analysis
# Gender distribution
sns.countplot(x='Gender', data=df)
plt.title("Gender Distribution")
plt.show()

# Gender vs Response
pd.crosstab(df['Gender'], df['Response']).plot(kind='bar', stacked=True)
plt.title("Response by Gender")
plt.ylabel("Count")
plt.show()
```

```python
#Vehicle Age and Claims
# Vehicle Age vs Response
pd.crosstab(df['Vehicle_Age'], df['Response']).plot(kind='bar', stacked=True, color=['red','green'])
plt.title("Response by Vehicle Age")
plt.ylabel("Count")
plt.show()
```

```python
#Region-wise Analysis
# Top 10 regions by response rate
region_resp = df.groupby('Region_Code')['Response'].mean()*100
top_regions = region_resp.sort_values(ascending=False).head(10)

plt.figure(figsize=(8,4))
top_regions.plot(kind='bar', color='teal')
plt.title("Top 10 Regions by Response Rate")
plt.ylabel("Response Rate (%)")
plt.show()
```

```python
#Policy Analysis
# Policy Sales Channel response rates (Top 10)
psc_resp = df.groupby('Policy_Sales_Channel')['Response'].mean()*100
top_channels = psc_resp.sort_values(ascending=False).head(10)

plt.figure(figsize=(8,4))
top_channels.plot(kind='bar', color='purple')
plt.title("Top 10 Policy Channels by Response Rate")
plt.ylabel("Response Rate (%)")
plt.show()
```

```python
#Claim Frequency by Vehicle Damage
# Vehicle Damage vs Response
pd.crosstab(df['Vehicle_Damage'], df['Response']).plot(kind='bar', stacked=True, color=['grey','green'])
plt.title("Response by Vehicle Damage")
plt.ylabel("Count")
plt.show()
```

```python
#Customer Loyalty (Number of Policies)
# Loyalty (Vintage) vs Response
plt.figure(figsize=(6,4))
sns.kdeplot(df[df['Response']==0]['Vintage'], label='No Response', shade=True)
sns.kdeplot(df[df['Response']==1]['Vintage'], label='Response', shade=True)
plt.title("Customer Loyalty (Vintage) vs Response")
plt.xlabel("Vintage (days)")
plt.legend()
plt.show()
```

```python
#Time Analysis (Temporal Patterns)
# Response rate by Vintage groups
df['Vintage_Group'] = pd.cut(df['Vintage'], bins=[0,50,100,150,200,250,300], labels=['0-50','51-100','101-150','151-200','201-250','251-300'])
vintage_resp = df.groupby('Vintage_Group')['Response'].mean()*100

vintage_resp.plot(kind='bar', color='brown')
plt.title("Response Rate by Customer Vintage Groups")
plt.ylabel("Response Rate (%)")
plt.show()
```
